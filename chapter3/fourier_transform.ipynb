```python
import cv2
import numpy as np
from scipy import ndimage
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("yoni_eyal.jpg")
```

```python
#pyplot expects RGB
def mymshow(image):
    fig, ax = plt.subplots()
    fig.set_size_inches(8,8)
    
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    ax.imshow(image)
```

```python
mymshow(img)
```

```python
# apply HPF
```

```python
#load image greyscale
img = cv2.imread("yoni_eyal.jpg", cv2.IMREAD_GRAYSCALE)

kernel_3x3 = np.array([[-1, -1, -1],
                       [-1, 8, -1],
                       [-1, -1, -1]])

kernel_5x5 = np.array([[-1, -1, -1, -1, -1],
                       [-1, 1, 2, 1, -1],
                       [-1, 2, 4, 2, -1],
                       [-1, 1, 2, 1, -1],
                       [-1, -1, -1, -1, -1]])

k3 = ndimage.convolve(img, kernel_3x3)
k5 = ndimage.convolve(img, kernel_5x5)

blurred = cv2.GaussianBlur(img, (117,117), 0)
g_hpf = img - blurred
```

```python
mymshow(k5)
```

```python
mymshow(g_hpf)
```
